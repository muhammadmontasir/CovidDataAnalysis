```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style(style='dark')
%matplotlib inline
```

```python
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
#confirmed.describe()
#confirmed.info()
#confirmed.dtypes
#confirmed.describe(include=object).transpose
#confirmed.isnull().any()
confirmed = confirmed.rename(columns= {"Province/State" : "state", "Country/Region" : "country"})

confirmed
```

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
deaths = deaths.rename(columns = {"Province/State" : "state", "Country/Region" : "country"})


deaths
#deaths.describe()
#deaths.info()
#deaths.dtypes
#deaths.describe(include=object).transpose
#deaths.isnull().any()
```

```python
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
recovered = recovered.rename(columns = {"Province/State" : "state", "Country/Region" : "country"})


recovered
#recovered.describe()
#recovered.info()
#recovered.dtypes
#recovered.describe(include=object).transpose
#recovered.isnull().any()
```

```python
f = plt.figure(figsize=(9,9), dpi = 100)
ax =f.add_subplot(111)
ax.set_axisbelow(True)

#tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area.
plt.tight_layout(pad= 10, h_pad = 5, w_pad = 5)

#plotting confirmed cases
#calculating data from the 48th column which represents 6th march. First covid case was reported in 8th march in Bangladesh
con_cases = np.sum(np.asarray(confirmed[confirmed['country'] == 'Bangladesh'].iloc[:, 48:]),axis = 0) #converting to column vector
date = np.arange(1, con_cases.shape[0] + 1)

plt.plot(date, con_cases, label = "Confirmed", color = 'darkslateblue', linewidth=2, linestyle= '--', marker = 'o', markersize = 3)
plt.fill_between(date, con_cases, color = 'darkslateblue', alpha = 0.1)

total_confirmed = np.sum(np.asarray(confirmed[confirmed['country'] == 'Bangladesh'].iloc[:, 48:]),axis = 0)[-1]

#plotting deaths cases
dth_cases = np.sum(np.asarray(deaths[deaths['country'] == 'Bangladesh'].iloc[:, 48:]), axis = 0) #converting to column vector

plt.plot(date, dth_cases, label = "Deaths", color= 'firebrick', linewidth = 2 , linestyle = '-', marker = 'o', markersize = 2)
plt.fill_between(date, dth_cases, color = 'firebrick', alpha = 0.2)
total_death = np.sum(np.asarray(deaths[deaths['country'] == 'Bangladesh'].iloc[:, 48:]), axis = 0)[-1] #counts total number of death from the vector

plt.text(0, 0.7, "Total Confirmed: " + str(total_confirmed) + "\n" + "Total deaths: " + str(total_death), fontsize = 15, horizontalalignment = 'left', verticalalignment = 'top', transform=ax.transAxes)
plt.legend(loc= "best", fontsize= 15)
plt.title('Covid Data Analysis : Confirmed cases and death cases(linear scale)')
plt.xlabel('No. of days')
plt.ylabel('No. of people')
plt.show()
```

```python
#using pie chart
labels = 'Active: '+ str(total_confirmed-(total_recovered + total_death)), 'Recovered: ' + str(total_recovered), 'Deaths: ' + str(total_death)
size = [total_confirmed, total_recovered, total_death]
explode = (0, 0, 0.3)
f3 = plt.figure(figsize=(7,7), dpi = 100)
ax3 =f3.add_subplot(111)

ax3.pie(size, explode=explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle= -120)
ax3.axis('equal')
plt.show()
```
